### What is the total Protein supply quantity in Madagascar in 2015?

```python
#to groupby the elements that each countries produced in the year 2015
x = df[['Area','Element', 'Y2015']].groupby(['Area', 'Element'])['Y2015'].sum()
x
```

```python
#to get the specific products from Madagascar
x2 = x.at['Madagascar']
x2
```

```python
#to get the specific protein supply quantity by madagascar in the year 2015
madagascar_protein_supply = x2.at['Protein supply quantity (g/capita/day)']
madagascar_protein_supply
```

```python
print('The specific protein supply quantity by madagascar in the year 2015 is :')
print(madagascar_protein_supply)
```


### Select columns ‘Y2017’ and ‘Area’, Perform a groupby operation on ‘Area’.  Which of these Areas had the 7th lowest sum in 2017?

```python
y = df[['Area', 'Y2017']].groupby('Area')['Y2017'].sum().reset_index()
y
```

```python
sorted_values = y.sort_values(by='Y2017')
sorted_values
```

```python
seventh_lowest_sum = sorted_values.iloc[6]
seventh_lowest_sum
```

```python
print("Area with the 7th lowest sum in 2017:")
print(seventh_lowest_sum)
```
